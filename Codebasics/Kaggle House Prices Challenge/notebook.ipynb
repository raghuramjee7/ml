```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib.inline
```

```python
data = pd.read_csv("train.csv")
```

```python
data = data.drop("Id",axis=1)
data.head()
```

```python
data.info()
```

```python
data.describe()
```

```python
data.select_dtypes(exclude=['object']).describe().transpose().round(decimals=2)
```

```python
data.select_dtypes(include=['object']).describe().transpose()
```

```python
data['SalePrice'].describe()
```

```python
sns.distplot(data['SalePrice'],hist=False,rug=True)
```

```python
sns.displot(np.log(data['SalePrice']))
```

```python
numerical_attributes = data.select_dtypes(exclude='object').drop(['SalePrice'],axis = 1).copy()
fig = plt.figure(figsize=(12,18))
for i in range(len(numerical_attributes.columns)):
    fig.add_subplot(9,4,i+1)
    sns.distplot(numerical_attributes.iloc[:,i].dropna(),hist=False,rug=True)
    plt.xlabel(numerical_attributes.columns[i])
plt.tight_layout()
plt.show()
```

```python
fig = plt.figure(figsize=(10,15))

for i in range(len(numerical_attributes.columns)):
    fig.add_subplot(9,4,i+1)
    sns.boxplot(y=numerical_attributes.iloc[:,i])
plt.tight_layout()
plt.show()
```
